```python
from aoc.utils import extract_coin
import skimage.io
import matplotlib.pyplot as plt
import numpy as np

image_path = "data/20241117_213835.jpg"
image = skimage.io.imread(image_path)

resize_shape = (200, 200)
bounding_box = (330, 940, 970, 1590)

image = extract_coin(image, bounding_box, resize_shape)

skimage.io.imshow(image)
plt.show()
```

```python
from aoc.utils import generate_rotations

rot_images = list(generate_rotations(image, 8))

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 15))

ax1.imshow(rot_images[0][0], cmap="gray")
ax1.set_title(f"{rot_images[0][1]} degrees")
ax2.imshow(rot_images[1][0], cmap="gray")
ax2.set_title(f"{rot_images[1][1]} degrees")
ax3.imshow(rot_images[2][0], cmap="gray")
ax3.set_title(f"{rot_images[2][1]} degrees")
ax4.imshow(rot_images[3][0], cmap="gray")
ax4.set_title(f"{rot_images[3][1]} degrees")

plt.show()
```
